```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import astropy.constants as const
import astropy.units as u
```

```python
nm = u.nm.to(u.cm)
micron = u.micron.to(u.cm)
kB = const.k_B.value
h = const.h.cgs.value
c = const.c.cgs.value
```

```python
# c1 = 3.742*10**8
# c2 = 1.439*10**4
T =6000 
WL_min = 1000*micron
WL_max = 10000*micron
num_points = 1000

wavelength = np.linspace(WL_min, WL_max, num_points)
# lm = np.linspace(0.001,1, num=50)

# E = c1/(lm**5*(np.exp(c2/(lm*T))-1))
```

```python
def B_lambda(lam, T):
    B_lambda = 2*h*c**2/lam**5*1/(h*c/(lam*kB*T))-1
    return B_lambda
```

```python
B_lambda_vals = B_lambda(wavelength,T)
```

```python
def B_peak_est (T):
    lam_peak = h*c/(kB*T)
    return lam_peak
```

```python
lam_peak_est = B_peak_est(T)
```

```python
plt.plot(wavelength/micron, B_lambda_vals)
plt.xlabel('wavelength(microns)')
plt.axvline('lam_peak_est/micron')
```

```python
plt.plot(wavelength/micron, B_lambda(wavelength,4000), color='blue')
plt.axvline(B_peak_est(4000)/micron, color='blue')
plt.plot(wavelength/micron, B_lambda(wavelength,5000), color='red')
plt.axvline(B_peak_est(5000)/micron, color='red')
plt.plot(wavelength/micron, B_lambda(wavelength,3000), color='green')
plt.axvline(B_peak_est(3000)/micron, color='green')
plt.plot(wavelength/micron, B_lambda(wavelength,2000), color='orange')
plt.axvline(B_peak_est(2000)/micron, color='orange')
plt.plot(wavelength/micron, B_lambda(wavelength,1000), color='black')
plt.axvline(B_peak_est(1000)/micron, color='black')
plt.xlabel('wavelength(microns)')
plt.yscale('log')
plt.ylim(1e7,1e15)
```
